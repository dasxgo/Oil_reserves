```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
field = 'IDARADO2'
```

## Upload WCA file


```python
df_wca = pd.read_csv('/home/dasxgo/developer/rock/WCA_2.csv')
data_major = 'OIL'
data_reservoir = 'WCA'
df_wca.insert(0, 'FIELD', field) 
df_wca.insert(1, 'MAJOR', data_major)
df_wca.insert(2, 'RESERVOIR', data_reservoir)
df_wca.head()
```

## Upload WCB file

```python
df_wcb = pd.read_csv('/home/dasxgo/developer/rock/WCB_2.csv')
data_major = 'OIL'
data_reservoir = 'WCB'
df_wcb.insert(0, 'FIELD', field) 
df_wcb.insert(1, 'MAJOR', data_major)
df_wcb.insert(2, 'RESERVOIR', data_reservoir)
df_wcb.head()
```

## Upload LSS file

```python
df_lss = pd.read_csv('/home/dasxgo/developer/rock/LSS_2.csv')
data_major = 'OIL'
data_reservoir = 'WCB'
df_lss.insert(0, 'FIELD', field) 
df_lss.insert(1, 'MAJOR', data_major)
df_lss.insert(2, 'RESERVOIR', data_reservoir)
df_lss.head()
```

## **Flatflie**

```python
df_fl = pd.concat([df_lss,df_wca,df_wcb], ignore_index=True)
df_fl['ProducingMonth'] = pd.to_datetime(df_fl['ProducingMonth'])
df_fl
```

## Data according to plan



```python
well_planned = 5000
data_oil = df_fl['LiquidsProd_BBL'] * (well_planned / df_fl['LateralLength_FT']) * (30.4 / df_fl['ProducingDays'])
data_gas = df_fl['GasProd_MCF'] * (well_planned / df_fl['LateralLength_FT']) * (30.4 / df_fl['ProducingDays'])
data_water = df_fl['WaterProd_BBL'] * (well_planned / df_fl['LateralLength_FT']) * (30.4 / df_fl['ProducingDays'])

df_fl['OIL 5K'] = data_oil
df_fl['GAS 5K'] = data_gas
df_fl['WATER 5K'] = data_water

df_fl['OIL 5K'] = np.array(df_fl['OIL 5K'], np.int64)
df_fl['GAS 5K'] = np.array(df_fl['GAS 5K'], np.int64)
df_fl['WATER 5K'] = np.array(df_fl['WATER 5K'], np.int64)

df_fl.head()
```
