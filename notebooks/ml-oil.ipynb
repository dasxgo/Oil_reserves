## **Flatflie**

```python
df_flatfile = pd.concat([df1,df2,df3], ignore_index=True)
df_flatfile['ProducingMonth'] = pd.to_datetime(df_flatfile['ProducingMonth'])
df_flatfile.head()
```

### Ajustando al plan



```python
def calculate_production_values(df_flatfile, well_planned):
    production_columns = ['LiquidsProd_BBL', 'GasProd_MCF', 'WaterProd_BBL']
    production_data = {}

    for column in production_columns:
        data = df_flatfile[column] * (well_planned / df_flatfile['LateralLength_FT']) * (30.4 / df_flatfile['ProducingDays'])
        production_data[column] = data

    df_flatfile[['OIL_BBL', 'GAS_MCF', 'WATER_BBL']] = pd.DataFrame(production_data)

    # Convert columns to int64
    for column in production_columns:
        df_flatfile[['OIL_BBL', 'GAS_MCF', 'WATER_BBL']] = np.array(df_flatfile[['OIL_BBL', 'GAS_MCF', 'WATER_BBL']], np.int64)

    return df_flatfile

well_planned = 5000
df_flatfile= calculate_production_values(df_flatfile, well_planned)
df_flatfile.head()
```

### Utilizando las demas variables

Un modelo de machine learning necesita variables numericas

```python
df = pd.get_dummies(df_flatfile, columns=['FIELD', 'MAJOR',	'RESERVOIR','API_UWI', 'ENVOperator',	'WellName', 'ProducingMonth'], drop_first=True)
df.head()
```

### Creando modelos

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Normaliza y realiza entrenamiento

```python
X_cols = list(set(df.columns)-set(['OIL_BBL']))
y_col = ['OIL_BBL']

X = df[X_cols].values
y = df[y_col].values

X_train, X_test, y_train, y_test = train_test_split(X,y)
sc_x = StandardScaler().fit(X)
sc_y = StandardScaler().fit(y)

X_train = sc_x.transform(X_train)
X_test = sc_x.transform(X_test)
y_train = sc_y.transform(y_train)
y_test = sc_y.transform(y_test)

model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

## Prediccion

```python
y_pred.shape
```

### Funciones metricas

El siguiente punto es crear las funciones que se utilizan para obtener las metricas del modelo

```python
import sklearn.metrics as metrics
mse = metrics.mean_squared_error(y_test,y_pred)
r2 = metrics.r2_score(y_test, y_pred)

print("r2 ", r2.round(4))
print("mse: ", mse.round(4))
```

## Entrenamiento de modelo de clasificacion con arbol de decision

```python
pip install category-encoders
```

```python
# Importante todos nuestros tipos de datos son object, realizamos una transformacion

import category_encoders as ce 

encoder = ce.OrdinalEncoder(cols = ['FIELD', 'MAJOR', 'RESERVOIR', 'API_UWI', 'ENVOperator', 'WellName'])

X_train = encoder.fit_transform(X_train)
X_test = encoder.fit_transform(X_test)
```

```python
# Verificamos la transformacion

X_train.head()
```

```python
# Importa arbol de decision

from sklearn.tree import DecisionTreeClassifier

# Creacion de modelo

tree = DecisionTreeClassifier(max_depth=3, random_state = 0)
```

```python
# Entrenamiento 

tree.fit(X_train, y_train)
```

```python
# Calculo de predicciones en Train y Test

y_train_pred_tree = tree.predict(X_train)
y_test_pred_tree = tree.predict(X_test)
```

```python
y_train_pred_tree
```

## evaluacion de modelo de clasificacion con arbol de decision

```python
# Calculo de metricas

from sklearn.metrics import accuracy_score

# Calculo el accuracy en train

train_accuracy_tree = accuracy_score(y_train, y_train_pred_tree)

# Calculo el accuracy en test

test_accuracy_tree = accuracy_score(y_test, y_test_pred_tree)

print('El accuracy en train es: ', train_accuracy_tree)
print('El accuracy en test es: ', test_accuracy_tree)

```

```python
# verificamos el feature importance

import seaborn as sns

importances = tree.feature_importances_
columns = X.columns

plt.figure(figsize=(14, 6))

sns.barplot(x=columns, y=importances, palette='bright', saturation=2.0, edgecolor = 'black', linewidth=2)
plt.xticks(rotation=45, horizontalalignment = 'right', fontweight = 'light', fontsize ='x-large')
plt.title('importancia de cada Feature')
plt.show
```
